### Question 2

Write a program that creates a database named phonebook.db. The database should have a table named Entries, with columns for a person’s name and phone number. Next, write a CRUD application that lets the user add rows to the Entries table, look up a person’s phone number, change a person’s phone number, and delete specified rows.

```python
import sqlite3

def create_database():
    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute('''Create table if not exists entries 
                (name TEXT, phone_number TEXT)''')
    conn.commit()
    conn.close()
    
def add_entry(name, phone_number):
    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute("Insert into entries (name, phone_number) values (?, ?)", (name, phone_number))
    conn.commit()
    conn.close()
    
def lookup(name):
    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute("Select phone_number from entries where name=?", (name,))
    result = c.fetchone()
    conn.close()
    if result:
        return result[0]
    else:
        return "Entry not found"
    
def update_phone_number(name, new_phone_number):
    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute("update entries set phone_number=? where name=?", (new_phone_number, name))
    conn.commit()
    conn.close()
    
def delete_entry(name):
    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute("Delete from entries where name=?", (name,))
    conn.commit()
    conn.close()
    
def main():
    create_database()
    while True:
        print("\nPhonebook Application")
        print("1. Add Entry")
        print("2. Lookup Entry")
        print("3. Update Phone Number")
        print("4. Delete Entry")
        print("5. Exit")
        choice = input("Enter your choice (1-5): ")
        
        if choice == "1":
            name = input("Enter the name: ")
            phone_number = input("Enter the phone number: ")
            add_entry(name, phone_number)
        elif choice == "2":
            name = input("Enter the name to lookup: ")
            result = lookup(name)
            print("Phone number: ", result)
        elif choice == "3":
            name = input("Enter the name whose phone number you want to update: ")
            new_phone_number = input("Enter the new phone number: ")
            update_phone_number(name, new_phone_number)
            print("Phone number updated")
        elif choice == "4":
            name = input("Enter the name of the entry to delete: ")
            delete_entry(name)
            print("Entry deleted")
        elif choice == "5":
            print("Exiting phonebook application")
            break
        else:
            print("Invalid choice")
            
if __name__ == "__main__":
    main()
```
